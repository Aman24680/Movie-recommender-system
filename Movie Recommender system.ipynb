```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('movies.csv')
movies_data.head()
```

```python
selected_feature = ['genres','keywords','tagline','cast','director']
for features in selected_feature:
  movies_data[features] = movies_data[features].fillna('')
```

```python
combined_features = movies_data['genres']+''+ movies_data['keywords']+''+ movies_data['tagline']+''+ movies_data['cast']+''+ movies_data['director']
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
featured_vector = vectorizer.fit_transform(combined_features)
print(featured_vector)
```

```python
similarity = cosine_similarity(featured_vector)
print(similarity)
```

```python
movie_name = input('Enter your movie : ')
```

```python
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
best_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(best_match)
```

```python
close_match = best_match[0]
print(close_match)
```

```python
index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
sorted_score = sorted(similarity_score, key = lambda x:x[1], reverse= True)
print(sorted_score)
```

```python
print("Recommended Movies :\n ")
i = 1
for movie in sorted_score:
  index = movie[0]
  movies_name = movies_data[movies_data.index == index]['title'].values[0]
  if (i<30):
    print(i,'.',movies_name)
    i=i+1
```
